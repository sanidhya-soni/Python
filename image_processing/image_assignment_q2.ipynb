```python
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mat_img
```

```python
def ChSwap(r, g, b):
    return Image.merge("RGB", (b, r, g))
```

```python
img_1 = Image.open(r"RGB.jpg")
img_1.show()
```

```python
r, g, b = img_1.split()
new_rgb = ChSwap(r, g, b)
new_rgb.show()
new_rgb.save('Create\RGBSwap.png', 'PNG')
```

```python
img_2 = mat_img.imread('Create\RGBSwap.png')
f = plt.figure()
f.add_subplot(1,2, 1)
plt.imshow(img_1)
f.add_subplot(1,2, 2)
plt.imshow(img_2)
```
